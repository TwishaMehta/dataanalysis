```python
sns.set_style("dark")

# Reshape the dataset from wide to long format
df_long = df.melt(id_vars=['ï»¿Countries and areas'], 
                  value_vars=['Completion_Rate_Primary_Male', 'Completion_Rate_Primary_Female', 'Completion_Rate_Lower_Secondary_Male', 'Completion_Rate_Lower_Secondary_Female', 'Completion_Rate_Upper_Secondary_Male', 'Completion_Rate_Upper_Secondary_Female'],
                  var_name='Education_Gender', value_name='Completion Rate')

# Split the 'Education_Gender' into two columns: 'Education Level' and 'Gender'
df_long[['Education Level', 'Gender']] = df_long['Education_Gender'].str.rsplit('_', 1, expand=True)

# Drop the original combined column
df_long.drop('Education_Gender', axis=1, inplace=True)

palette = {'Male': 'blue', 'Female': 'pink'}

# Create a FacetGrid, using country as the facet
g = sns.FacetGrid(df_long, col='ï»¿Countries and areas', col_wrap=7, height=2.5, sharey=True)

# Map the barplot to each facet
g.map(sns.barplot, 'Education Level', 'Completion Rate', 'Gender', order=['Completion_Rate_Primary', 'Completion_Rate_Lower_Secondary', 'Completion_Rate_Upper_Secondary'], palette=palette)

# Additional plot formatting
g.set_xticklabels(rotation=45)
g.set_titles('{col_name}')
g.fig.suptitle('Education Completion Rates by Country, Gender, and Education Level', fontsize=16, y=1)

plt.subplots_adjust(top=1)
g.add_legend(title="Gender", adjust_subtitles=True)


# Adjust layout
plt.tight_layout()
plt.show()
```
